```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

```python
df=pd.read_csv("demographic-statistics.csv", sep=";")
#print(df)
df.rename(columns = {'anna_c_e':'year', 'da_c_clarations_de_naissance':'births',
                     'da_c_clarations_de_mariage':'marriages','mortinatalita_c':'stillbirth'}, inplace = True)

df["year"]= pd.to_datetime(df["year"], format='%Y')

df=df.sort_values(by='year')
df
```

```python
fig,ax= plt.subplots()
ax.plot(df['year'], df['marriages'], color='red', label="marriages", marker='o')
ax.plot(df['year'], df['divorces'], color='green', label="divorces", marker='s', linestyle="--")

ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title('Marriages and divorces', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('quantity', fontsize=14)
ax.spines[['right', 'top']].set_visible(False)

plt.gcf().autofmt_xdate()
ax.legend()
plt.show()
```

```python
fig,ax= plt.subplots()
ax.plot(df['year'], df['births'], color='red', label="marriages", marker='o')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title('Births by year', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Births', fontsize=14)
ax.spines[['right', 'top']].set_visible(False)

plt.gcf().autofmt_xdate()

plt.show()
```
