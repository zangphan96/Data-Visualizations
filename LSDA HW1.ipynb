```python
import numpy as np
from scipy.stats import ortho_group
from numpy.linalg import norm
from matplotlib import pyplot
import math

B=np.random.random((100, 100)) #random matrix


A=np.matmul(B.T,B) #should be 100*100
#print(A)

b=A[:,0]+A[:,1]+A[:,2] #should be 100*1
#print(b)

```

```python
def objective(x):
    return np.matmul(np.matmul(x.T,A),x)-np.matmul(b.T,x)
    
def derivative(x):
    return 0.5*np.matmul(A.T,x)+0.5*np.matmul(A,x)-b
```

```python

def gradient_descent(objective, derivative, start, learn_rate, n_iter, tolerance):
     # track all solutions
    solutions, scores = list(), list()
    solution = start
    for i in range(n_iter):
        #calculate gradient
        gradient=derivative(solution)
        if (norm(gradient,2) <= tolerance):
            break
        solution = solution - learn_rate * gradient
        solution_eval= objective(solution)
        
         # store solution
        solutions.append(solution)
        scores.append(solution_eval)
        
         # report progress
        print('>%d f(%s) = %.5f' % (i, solution, solution_eval)) #value at each iteration
        
    return [solutions,scores,solution]

#implementation
learn_rate=0.2
n_iter = 100
tolerance=10e-5

start = np.random.uniform(-1,1,(100,)) #starting point
# perform the gradient descent search
solutions, scores, solution = gradient_descent(objective, derivative, start, learn_rate, n_iter, tolerance)
```

```python
#matrix of optimal solutions
print(solution)
#opt_matrix=np.full((100, 1), solution)
print(solution.shape)
```

```python
opt_matrix=np.full((100, 100), solution)
```

```python
differences=math.log(norm(solutions-opt_matrix,2, 1))
print(differences.shape)
#gradient descent iterations
iterations=np.arange(1,len(solutions)+1)
# plot the solutions found
pyplot.plot(iterations, differences, '.-', color='red')
# show the plot
pyplot.show()
```
