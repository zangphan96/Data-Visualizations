```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df=pd.read_csv("BTCvsSP500.csv")
print(df)
df["date"]= pd.to_datetime(df["date"])
df.info()
```

```python
#original and Jorma graph
fig,ax= plt.subplots()
ax.plot(df['date'], df['BTC'], color='red', label="BTC")
ax.plot(df['date'], df['SP500'], color='green', label="SP500")

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))

plt.title('Cryptocurrencies performance', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('Closing price', fontsize=14)
ax.spines[['right', 'top']].set_visible(False)

plt.gcf().autofmt_xdate()
ax.legend()
plt.show()
```

```python
#Bitcoin-tailored graph
df_BTC=df[(df['date'] > '2015-01-01') & (df['date'] < '2017-12-07')]
print(df_BTC)
```
